```python
def get_time(data):
  time=data.split(" ")
  hour=time[1].split(":")[0]
  return hour
df['time']=df["trans_date_trans_time"].apply(get_time)
```

```python
df.head()
```

```python
df.drop(['trans_date_trans_time'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
num=LabelEncoder()
df['merchant']=num.fit_transform(df['merchant'])
```

```python
df['category']=num.fit_transform(df['category'])
df['city']=num.fit_transform(df['city'])
df['state']=num.fit_transform(df['state'])
df['city_pop']=num.fit_transform(df['city_pop'])
df['job']=num.fit_transform(df['job'])
df['time']=num.fit_transform(df['time'])
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
column=["cc_num","merchant","category","amt","city","state","zip","city_pop","job","unix_time","merch_long","time"]
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
df[column]=scale.fit_transform(df[column])
```

```python
df.head()
```

```python
df['is_fraud']
```

```python
x_train=df.drop("is_fraud",axis=1)
```

```python
y_train=df["is_fraud"]
```

```python
y_train
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
train_pred=model.predict(x_train)
```

```python
print("Logistic Regression:")
print("Accuracy:", accuracy_score(y_train, train_pred))
print(classification_report(y_train, train_pred))
```
